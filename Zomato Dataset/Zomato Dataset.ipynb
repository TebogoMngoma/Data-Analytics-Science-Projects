## Data Cleaning

```python
#drop any null values in our dataset
data.dropna(axis=0,inplace=True)
```

```python
#when can see that our data no longer contain null values
data.isnull().sum()
```

```python
#convert the approx_cost feature from catagorical to numerical
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].str.replace('\W', '', regex=True)
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(int)
```

```python
#Make copy of our dataset
data_df = data.copy()
```

```python
# Remove any redundant features(columns)
data_df = data_df.drop(columns=['phone', 'url','reviews_list','menu_item','address','dish_liked','cuisines'],axis=1)
```

```python
#converting rate feature to numerical
def cleaned_rate(value):

    if(value=='NEW' or value=='-'):
        return np.nan

    else:
        value=str(value).split('/')
        value=value[0]
        # return it by converting it from string to float
        return float(value)
    
data_df['rate']=data_df['rate'].apply(cleaned_rate)
```

```python
#converting rates to percantages 
data_df['rate'] = data_df['rate'] / 5.0 * 100
```

```python
#renaming feature names
data_df.rename(columns = {'name':'restuarant_name'}, inplace = True)
data_df.rename(columns = {'approx_cost(for two people)':'Avg_cost_for_two_people'}, inplace = True)
data_df.rename(columns = {'cuisines':'cuisines_served'}, inplace = True)
data_df.rename(columns = {'rest_type':'restuarant_type'}, inplace = True)
data_df.rename(columns = {'listed_in(type)':'service_type'}, inplace = True)

```

## Data Visualisation

### 1. Restuarants delivering online or not

```python
plt.figure(figsize=(6,6))
sns.countplot(x='online_order', data=data_df)
```

### 2. Restuarants allowing table booking or not

```python
plt.figure(figsize=(6,6))
sns.countplot(x='book_table', data=data_df)
```

### 3. Restuarant type 

```python
plt.figure(figsize=(16,6))
sns.countplot(x='restuarant_type', data=data_df,order=data_df['restuarant_type'].value_counts().iloc[:20].index)
plt.title('RESTUARANT TYPE', fontsize=16)
plt.xticks(rotation=90)

#add axis labels
plt.xlabel('Restuarant Type',fontsize=16)
plt.ylabel('Number Of Each Restuarant Type',fontsize=16)
```

### 4. Relation between Location and Rating


```python
grouped_location = data_df.groupby('location',as_index=False)['rate'].agg('mean')
```

```python
x_grouped_location = grouped_location['location']
```

```python
y_grouped_location = grouped_location['rate']
```

```python
plt.figure(figsize=(20,8))
sns.barplot(x = x_grouped_location,
            y = y_grouped_location,order=data_df['location'].value_counts().iloc[:48].index)
plt.xticks(rotation=90)
plt.title('AVERAGE RATING OF EACH LOCATION', fontsize=16)

# Show the plot
plt.xlabel('Location',fontsize=13)
plt.ylabel('Average Rating (%)',fontsize=13)
plt.show()
```
